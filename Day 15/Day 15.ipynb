```python
# ML:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Machine learning:
from sklearn import *
print('Done') 
```

```python
data = {'Study_hrs':range(1,11),
       'Marks Obtained':[25,33,46,50,55,67,78,85,95,100]}
df = pd.DataFrame(data)
df
```

```python
# ML model build?: x, y: +ve, -ve
# Scatter Plot
plt.title('Study vs Marks')
plt.scatter(df['Study_hrs'],df['Marks Obtained'],marker = 'o', color = 'r')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.show()
```

```python
# Import Module
from sklearn.linear_model import LinearRegression
# Object Model Creation
model = LinearRegression()
# X, y
X = df[['Study_hrs']]
y = df['Marks Obtained']
# Training/Learning
model.fit(X,y)
```

```python
x = [[5.5]]
m = model.coef_
c = model.intercept_
print(m,c)
```

```python
std_hrs = [[5.5]]
marks = m * std_hrs + c
print('Predicted Marks:',marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs = input('Enter study hours:')
std_hrs = [[float(study_hrs)]]
ans = model.predict(std_hrs)
marks = round(ans[0],2)
print(f'Your predicted marks is: {marks}')
```
