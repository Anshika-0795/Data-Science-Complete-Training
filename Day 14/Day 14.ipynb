```python
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex')  # hue: further divide: based on categorical(object)
plt.show()
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex',hue = 'alive',palette = sns.color_palette('magma',2))  # hue: further divide: based on categorical(object)
plt.title('Tiatanic Survival Analysis Based On Gender')
plt.show()
```

```python
# sns.barplot
sns.barplot(data = df_titanic, x= 'class', y = 'fare',hue = 'alive',estimator='mean') # min, max, sum
plt.show()
# Confidence Interval(CI): 95%
```

```python
# sns.barplot
sns.barplot(data = df_titanic, x= 'class', y = 'fare',hue = 'alive',errorbar = ('ci',0)) # min, max, sum
plt.grid(color = 'r',alpha = 0.1)
plt.show()
# Confidence Interval(CI): 95%
```

```python
# # sns.boxplot: To detect Outliers and to check data distribution
# # To detect Outliers:
# # Range: Min to max(19 - 22): 35, 28, 45(Outliers: fliers
# # Q2: Quantile : 50%
# # Q1: 25%
# # Q3: 75%
# # IQR = Q3 - Q1
# UW(Upper cap) = Q3 + 1.5*IQR
# LW(Lower cap) = Q1 - 1.5* IQR
```

```python
# sns.boxplot
sns.boxplot(data = df_titanic, x='class', y   = 'fare',hue = 'sex' )
plt.show()
```

```python
# sns.boxplot
sns.boxplot(data = df_titanic, y = 'class', x = 'fare',hue = 'sex',orient = 'h' )
plt.show()
```

```python
# sns.boxplot
plt.figure(figsize = (10,5))
sns.boxplot(data = df_titanic, y='class', x   = 'fare' , hue = 'sex',orient = 'h')
plt.show()
```

```python
# sns.violin plot
plt.figure(figsize = (10,5))
sns.violinplot(data = df_titanic, x='class', y   = 'fare' , hue = 'sex',orient = 'v')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
# sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
# sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
```

```python
# sns.kdeplot : Kernel density estimation

sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width',kind = 'hex') # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```
